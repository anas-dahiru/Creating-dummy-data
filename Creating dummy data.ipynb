[Problem 4] Addition of data

```python
import numpy as np
#create the variables
mean = [-0, -3]
cov = [[1.0, 0.8], [0.8, 1.0]]
np.random.seed(0)
rand_num1 = np.random.multivariate_normal(mean, cov, 500)
print(rand_num)
rand_num.shape
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
#plot.figure(figsize=(20,20))
plt.title("Two in one scatter plot")
plt.xlabel("x1")
plt.ylabel("x2")

plt.scatter(rand_num[:,0], rand_num[:,1],s=20, label='0', marker='o')

plt.scatter(rand_num1[:,0], rand_num1[:,1], s=20,label='1', marker='o')
plt.legend()
plt.show()
```

[Problem 5] Data combination

```python
data_combine = np.concatenate((rand_num, rand_num1),axis=0)
print(data_combine)
data_combine.shape
```

using v stack




```python
combined_data = np.vstack((rand_num, rand_num1))
combined_data.shape
```
